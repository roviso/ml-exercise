```python
import pandas as pd

```

```python
def combine_files(csv1, csv2):
    # Read the CSV files
    df1 = pd.read_csv(csv1, header=None)
    df2 = pd.read_csv(csv2, header=None)

    # Merge the DataFrames on the first column
    merged_df = pd.merge(df1, df2, on=0)

    # Set the first column as the index
    merged_df.set_index(0, inplace=True)


    # Write the result to the output file
    
    return df1, df2, merged_df

```

```python
# Call the function with your file paths
df1, df2, merged_df = combine_files('Java/resources/1.csv', 'Java/resources/2.csv')

outputFileName = "output.csv"
merged_df.to_csv(outputFileName, index=False, header=False)
```
